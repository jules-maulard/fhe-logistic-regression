# Sigmoid approximation exploration

```python
import numpy as np
import matplotlib.pyplot as plt

import sys
import os
sys.path.append(os.path.abspath('..'))
from src.config import setup_logging
logger = setup_logging("DEBUG")

from src.sigmoid import get_sigmoid_approx
```

## Least squares

```python
sigmoid_true = get_sigmoid_approx()
minus_sigmoid = get_sigmoid_approx("minus_sigmoid")
lsq3 = get_sigmoid_approx("least_squares", 3)
lsq5 = get_sigmoid_approx("least_squares", 5)
lsq7 = get_sigmoid_approx("least_squares", 7)
```

```python
x = np.linspace(-10, 10, 1000)

y_true = sigmoid_true(x)
y_minus = minus_sigmoid(x)
y_lsq3 = lsq3(x)
y_lsq5 = lsq5(x)
y_lsq7 = lsq7(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_lsq3, label="LSQ3")
plt.plot(x, y_lsq5, label="LSQ5")
plt.plot(x, y_lsq7, label="LSQ7")
plt.plot(x, y_minus, label="Sigmoid (-x)", linestyle="--")
plt.plot(x, y_true, label="Sigmoid (x)", linestyle="--")

plt.title("Real sigmoid and approximations")
plt.ylim(-0.2, 1.2)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(x, np.abs(y_minus - y_lsq3), label="LSQ3 Error")
plt.plot(x, np.abs(y_minus - y_lsq5), label="LSQ5 Error")
plt.plot(x, np.abs(y_minus - y_lsq7), label="LSQ7 Error")
plt.title("Approximations errors regarding real sigmoid")
plt.ylim(-0.01, 0.17)
plt.legend()
plt.grid()
plt.show()
```
